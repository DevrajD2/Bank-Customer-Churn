Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing Data

```python
df = pd.read_csv('Bank Churn.csv')
df
```

```python
ax = sns.countplot(x = 'Churn', data = df)

ax.bar_label(ax.containers[0])
plt.title("No. of Customers Churned")
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Customers Churned", fontsize = 10)
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "Geography", data = df, hue = "Churn" )
ax.bar_label(ax.containers[0])
plt.title("Churned per Location")
plt.show()
```

```python
plt.figure(figsize = (3,3))
sns.countplot(x = "Gender", data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
df.columns.values
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "IsActiveMember", data = df,)
ax.bar_label(ax.containers[0])
plt.title("Current active members")
plt.show()
```

```python
plt.figure(figsize = (9,4))
sns.histplot(x = "EstimatedSalary", data = df, bins = 72,)
plt.show()
```

```python
total_counts = df.groupby('Card Type')['Churn'].value_counts(normalize=True).unstack() * 100
fig, ax = plt.subplots(figsize=(4, 4))  
total_counts.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e'])  # Customize colors if desired

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center')

plt.title('Churn by Card Type (Stacked Bar Chart)')
plt.xlabel('Card Type')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Churn', bbox_to_anchor = (0.9,0.9))  

plt.show()
```
